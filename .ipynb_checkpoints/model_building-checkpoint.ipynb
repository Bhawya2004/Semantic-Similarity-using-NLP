# Semantic Similarity Measurement
## Using Sentence Transformers and Cosine Similarity

This notebook demonstrates how to measure the **semantic similarity** between two sentences using:
- **Sentence Transformers** — to convert sentences into numerical vectors (embeddings)
- **Cosine Similarity** — to measure the similarity between those vectors

---

## Step 1 – Import Libraries

We need three libraries:
- `sentence_transformers` — provides pretrained models to generate sentence embeddings
- `sklearn` — provides the cosine similarity function
- `numpy` — for numerical operations

```python
# Import required libraries
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

print("Libraries imported successfully!")
```

## Step 2 – Load Pretrained Sentence Transformer Model

We use the **`all-MiniLM-L6-v2`** model from Sentence Transformers.

### What does this model do?
- It converts any sentence into a **384-dimensional numerical vector** (called an **embedding**).
- These embeddings capture the **semantic meaning** of the sentence.
- Sentences with similar meanings will have similar embeddings.

This is a **pretrained model** — it has already been trained on millions of sentence pairs, so we don't need to train it ourselves.

```python
# Load the pretrained Sentence Transformer model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

print("Model loaded successfully!")
print(f"Model name: all-MiniLM-L6-v2")
print(f"Embedding dimension: 384")
```

## Step 3 – Define Example Sentences and Generate Embeddings

We will take two example sentences and convert them into embeddings.

### What are embeddings?
- An **embedding** is a list of numbers (a vector) that represents the meaning of a sentence.
- For example, `"I love machine learning"` might become `[0.12, -0.45, 0.78, ...]` (384 numbers).
- Similar sentences will have similar embeddings.

```python
# Define two example sentences
sentence1 = "I love machine learning"
sentence2 = "I enjoy studying artificial intelligence"

print(f"Sentence 1: {sentence1}")
print(f"Sentence 2: {sentence2}")
print()
```

```python
# Generate embeddings for both sentences
# The model.encode() method converts sentences into numerical vectors
embeddings = model.encode([sentence1, sentence2])

print(f"Shape of embeddings: {embeddings.shape}")
print(f"  → 2 sentences, each represented as a 384-dimensional vector")
print()
print(f"Embedding for Sentence 1 (first 10 values): {np.round(embeddings[0][:10], 4)}")
print(f"Embedding for Sentence 2 (first 10 values): {np.round(embeddings[1][:10], 4)}")
```

## Step 4 – Compute Cosine Similarity

Now we measure how similar the two sentence embeddings are using **Cosine Similarity**.

### What is Cosine Similarity?
- It measures the **angle** between two vectors.
- If two vectors point in the same direction → the sentences have similar meanings.

### Score Range:
| Score | Meaning |
|-------|--------|
| **1** | Very similar (same meaning) |
| **0** | Unrelated |
| **-1** | Opposite meaning |

```python
# Compute Cosine Similarity between the two embeddings
similarity_score = cosine_similarity(
    [embeddings[0]],   # Embedding of Sentence 1
    [embeddings[1]]    # Embedding of Sentence 2
)

# Extract the score (it's a 2D array, so we get [0][0])
score = round(float(similarity_score[0][0]), 4)

print(f"Cosine Similarity Matrix:\n{similarity_score}")
print(f"\nSimilarity Score: {score}")
```

## Step 5 – Display and Interpret the Result

```python
# Display the final result
print("=" * 50)
print("       SEMANTIC SIMILARITY RESULT")
print("=" * 50)
print(f"Sentence 1: {sentence1}")
print(f"Sentence 2: {sentence2}")
print(f"\nSimilarity Score: {score}")
print()

# Interpret the score
if score >= 0.8:
    print("Interpretation: The sentences are HIGHLY SIMILAR in meaning.")
elif score >= 0.5:
    print("Interpretation: The sentences are MODERATELY SIMILAR in meaning.")
else:
    print("Interpretation: The sentences are NOT VERY SIMILAR in meaning.")

print("=" * 50)
```

## Bonus – Try More Sentence Pairs

Let's test with a few more examples to better understand how similarity works.

```python
# More example sentence pairs
test_pairs = [
    ("The cat is sitting on the mat", "A cat rests on a rug"),
    ("It is raining outside", "The weather is sunny and warm"),
    ("Python is a programming language", "Python is a type of snake"),
    ("I want to eat pizza", "I want to eat pizza"),
]

print("Similarity Scores for Multiple Sentence Pairs:")
print("=" * 60)

for s1, s2 in test_pairs:
    emb = model.encode([s1, s2])
    sim = cosine_similarity([emb[0]], [emb[1]])[0][0]
    print(f"\nSentence 1: {s1}")
    print(f"Sentence 2: {s2}")
    print(f"Similarity Score: {round(sim, 4)}")
    print("-" * 60)
```

## Summary

In this notebook, we:

1. **Loaded** a pretrained Sentence Transformer model (`all-MiniLM-L6-v2`)
2. **Generated embeddings** — numerical vectors representing the meaning of sentences
3. **Computed Cosine Similarity** — measured how close the embeddings are
4. **Interpreted the result** — a higher score means the sentences are more semantically similar

This approach is widely used in NLP for tasks like:
- Duplicate question detection
- Semantic search
- Plagiarism detection
- Text clustering
